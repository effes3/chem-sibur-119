```python
!pip install rdkit 

from rdkit import Chem, RDLogger
from rdkit.Chem import AllChem
RDLogger.DisableLog('rdApp.*')

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
to_clean_df = pd.read_csv('/kaggle/input/dataforclean/sibur_element_119_final_train_data80.csv')
to_clean_df = to_clean_df.drop(columns=['ID'])
to_clean_df
```

```python
train_cleaned = to_clean_df[to_clean_df['SMILES'].apply(lambda smi: Chem.MolFromSmiles(smi) is not None)] #чистим датасет от невалидных молекул
print(f'Было {len(to_clean_df)} молекул, почистили невалидные SMILES, и осталось {len(train_cleaned)}')
```

```python
train_cleaned = (
    train_cleaned
    .assign(SMILES=lambda x: x['SMILES'].str.strip())
    .groupby('SMILES', as_index=False)
    .agg({'LogP': 'mean'})
)
print(f'Дубликаты удалены. Осталось {len(train_cleaned)} уникальных молекул')
```

```python
print(f'Осталось {len(train_cleaned)} молекул, теперь у нас есть чистый датасет, сохраним его')
train_cleaned.to_csv('train_cleaned.csv', index=False)
```
